#### Bosons

The SM describes the electromagnetic force, the weak nuclear force, and the strong nuclear force, but is yet to describe the gravitational force. Bosons are fundamental particles that either have spin-1 or spin-0. They mediate the interactions between particles and act as force carriers. The electromagnetic force carrier is the photon (spin-1). The weak nuclear force carriers are the W and Z bosons (spin-1). The strong nuclear force carrier is the gluon (spin-1). The last boson, the Higgs, does not describe any fundamental force of nature. The Higgs boson describes how particles acquire mass through the interaction of the Higgs field. Additionally, at high enough energies, the electromagnetic and weak force merge into the electroweak force.

#### Fermions

Fermions are fundamental particles with spin-1/2 and consist of leptons and quarks. Leptons consist of "leptons" and neutrinos, where the "leptons" are electrons, muons, and taus. Each "lepton" has an electrically neutral counterpart, called neutrinos, namely, electron neutrino, muon neutrino, and tau neutrino.  Quarks consist of six flavours, namely, up, down, charm, strange, bottom (or beauty), and top. There are three up-type quarks, namely, up, charm, and top, and they are so named because they have positive electric charge. There are also three down-type quarks, namely, down, strange, and bottom, because they have negative electric charge. Quarks have an additional charge called "colour", and there are three types of colour charge, namely, blue, green, and red. 

There are three generations of fermions, where the first generation is responsible for all matter you see around you, and each suuccessive generation of particles is heavier than the previous generations.

The strong nuclear force says that quarks cannot exist in isolation, i.e., they always exist in pairs. However, the strong nuclear force also says that quarks appear to be free at high enough energies, known as asymptotic freedom. Additionally, the strong nuclear force says that free quarks and gluons combine to form composite particles called hadrons, also called hadronization. This hadronization time occurs within 10^-23 seconds. 

All the fundamental particles in the SM have an anti-particle counterpart which is identical in every way except that they posses the opposite electrical charge.

#### Particle Interaction

According to the SM, there are only few types of interactions that are allowed. All electrically charged particles can interact via the photon. Since the photon itself is electrically neutral, it does not interact with itself. All colour charged particles can interact via the gluon. Since the gluon itself is colour charged, it can interact with itself. All fermions possess the weak charge and therefore can interact via the W or Z bosons. Since the W boson carries the weak charge, it can interact with itself. The Z boson does not carry the weak charge, so it cannot interact with itself. 

The photon and Z boson has no structural impact on the particle, e.g., an electron to an electron. The gluon has no structural impact on the particle, but does change its colour charge, e.g., green top quark to anti-blue top quark. The W boson is the only boson interaction that changes the flavour of the particle. The W boson changes an up-type quark into a down-type quark, or vice versa, e.g., an top quark into a bottom quark. Additionally, the W boson changes a lepton into its corresponding lepton-neutrino e.g., electron to electron neutrino.

#### Particle Decay

All unstable particles will decay. All non-fundamental particles will split into their internal constituents when they decay. However, all fundamental particles have no internal constituents, so when they decay, they will transform all their energy and mass into new lighter particles. The allowed SM interactions also tell us how particles decay. Since the photon, gluon, and Z boson do not change the flavour of the particle - do not transform into new particles - they cannot be used to represent decays. Therefore, the W boson represents particle decays. However, due to the fact that the Z boson has mass, it can also decay.

Additionally, particles can annihilate to form other particles. A fermion interacting with its corresponding anti-fermion, will produce a boson. Furthermore,photons and gluons can undergo pair production, where the photon produces a lepton and its corresponding anti-lepton, while the gluon produces a quark and its corresponding anti-quark.

### The Top Quark

The top quark has an electric charge of +2/3, spin of 1/2, and a mass of approximately 172.5 GeV. It interacts with all fundamental forces of nature and is the heaviest fundamental particle in the SM.

#### Top Quark Production

Based on the allowed interactions of the SM, the top quark can be produced in a variety of ways. However, due to its mass, the production is highly dependent on the energy and type of the experiment being conducted. 

In electron-positron collisions, extremely high energies are needed to produce a top quark. These particles can only interact via the photon and the only way a photon can produce a top quark is to pair produce into a pair of top quarks. The photon has no mass and therefore all of its energy must be a minimum of more than twice the mass of the top quark. Increasing the energy of an electron in a circular collider brings about synchrotron radiation, which is inversely proportional of the particle's mass to the fourth power, and for light particles such as electrons, this is highly problematic.

However, the energy of a proton can be increased in a circular collider with less problematic synchrotron radiation, and therefore, we can produce top quarks. Since the quarks and gluons inside the proton interact in proton-proton collisions, the dominant production mode of a top quark is actually top-antitop quark pairs. These productions modes include gluon-gluon fusion and quark-antiquark annihilation.

Gluon-gluon fusion is when two gluons interact directly to form a top and an antitop pair. This accounts for approximately 85% of the top-antitop pairs in proton-proton collisions. Quark-antiquark annihilation is when two identical quarks with opposite electric charge annihilate to form a gluon and the gluon then decays into a top-antitop quark pair. #### Top Quark Decay

The top quark decays approximately 100% of the time to a W boson and bottom quark (also called b quark), i.e. 

$${t \rightarrow Wb}$$

The W boson decays 67% of the time hadronically - into two different flavours of quarks - and 33% of the time leptonically - into a lepton and its corresponding lepton neutrino. The W boson leptonic decay is split approximately equally among the three generation of leptons, i.e., 11% into an electron-electron neutrino, 11% into muon-muon neutrino, and 11% into a tau-tau neutrino.

$${W \rightarrow qq' \text{  or  } W \rightarrow l\nu}$$

The bottom quark will from hadrons through the process of hadronization and these hadrons will then decay.

#### Top Quark Importance

The top quark is not only the heaviest quark in the SM, it is also the heaviest fundamental particle in the SM. It is even heavier than the particle associated with giving particles their mass, i.e., the Higgs boson (m(H) = 125 GeV), and therefore, has the strongest interaction with the Higgs boson. The size of the top quark mass brings about a decay time of 10^-25 seconds. This decay time is faster than the time it takes for quarks to undergo hadronization (i.e., 10^-23 seconds), and therefore the top quark will decay before hadronization. As the top quark decays, it will directly pass on its kinematics and properties, such as its momentum and spin. Therefore, the properties of the top quark can be studied directly from its decay products. When performing precision measurements on top quark properties and interactions, any deviations from the SM theory could indicate signs of new physics. 

The top quark is also linked to various theories that go beyond the Standard Model, called BSM theories. An example of such theories is Supersymmetry.

### The Large Hadron Collider

The Large Hadron Collider (LHC) is a 27 km circular ring that spans across France and Switzerland. It consists of four major experiments with detectors situated along the ring. These four experiments are ATLAS, CMS, ALICE, and LHCb. The ATLAS and CMS detectors were designed as general purpose detectors to perform a wide variety of studies. ALICE, however, was designed to study heavy-ion collisions and the quark-gluon plasma, while LHCb was designed to study heavy-flavour hadrons, specifically bottom quarks - hence the "b" in the name. 

The large hadron collider is the largest circular collider in the world and is the last accelerator in the chain before the protons collide. A hydrogen gas bottle releases hydrogen atoms. An electric field is used to knock out the electron from the hydrogen atom, which leaves behind the proton. These protons are fed into multiple accelerators to speed them up to nearly the speed of light with energies of 6.5 TeV. These protons are bunched up together before colliding at the centre of the four major experiments. The aftermath of these collisions are then captured by the detectors and the data is stored for further analysis.

### The ATLAS Detector

The ATLAS detector is the largest detector in the world and is cylindrical in shape, which is about 45 metres long and 25 metres in diameter. Many subdetectors make the ATLAS detector starting from the Inner Detector, which is the closest subdetector to the collision point, all the way to the Muon Spectrometer, which is the outermost subdetector. Additionally, large magnets are used to focus the proton bunches and to help identify particles that come from the collision.

The Inner Detector (ID) is used to measure the trajectory of the particles that come from the collisions. These trajectories are reconstructed from the signals that are left in the detector as the particles interact with the material of the sensors. The next subdetector is the Electric Calorimeter (ECal) which is used to measure the energy of electromagnetically particles, specifically, electrons and photons. The Hadronic Calorimeter (HCal) follows and is used to measure the energy of hadrons. The last subdetector is the Muon Spectrometer (MS) which is used to measure the tracks and energy of muons.

### Experimental Signature

Once a top quark is produced, it will decay into a W boson and a bottom quark (also a b quark) approximately 100% of the time. Since we want as much top quarks as possible and the fact that top quark pair production is the dominant production mode in proton-proton collision, we will search for top quark pairs. A high-level view of the decay mode will be 

$${t\bar{t} \rightarrow W^+bW^-\bar{b}}$$

The W boson will decay while the b quark will undergo hadronization. There are numerous decay modes that can happen and therefore, we need to pick which decay mode to perform the search. Let's say we pick the decay mode with the highest decay ratio, i.e., both W bosons decayed hadronically, as a W boson decays hadronically about 67% of the time. This means our signature should be 

$${t\bar{t} \rightarrow W^+bW^-\bar{b}} \rightarrow (q_{1}q_{2})(q_{3}q_{4})b\bar{b}$$

This means we are looking for a signature of at least 6 jets (i.e., 2+2+2). This signature has both theoretical and experimental advantages and major disadvantages. Due to these disadvantages, we instead perform the search where one or both of the W bosons decay leptonically, i.e.

$${t\bar{t} \rightarrow W^+bW^-\bar{b}} \rightarrow (l\nu)(q_{1}q_{2})b\bar{b}$$

or

$${t\bar{t} \rightarrow W^+bW^-\bar{b}} \rightarrow (l_{1}\nu_{1})(l_{2}\nu_{2})b\bar{b}$$

Again, we have to choose one, but which one and why? Most analysis will already have chosen the decay mode based on the theoretical and experimental advantages and disadvantages, but let's use the data and then determine from there.

### Preselection

The data collected and simulated consists of a large number of events. We are potentially talking tens of millions and more, sometimes even billions. Imagine we conduct this analysis over all of those events. That seems normal and fine, but then imagine we want to study the impact of changing the number of leptons, or increasing the momentum selection of an object. This would take a very long time to perform these studies. Therefore, we perform a preselection to speed up the analysis.

Since we know we want a minimum of one lepton, and in this case we choose a minimum of one electron or muon, we can already apply a minimum selection of at least one lepton. This will automatically throw away all events with no leptons, greatly reducing the number of events to run over. Additionally, we know that the W boson also decays into a neutrino, and therefore we can apply a selection on the neutrinos. However, what selection? We know neutrinos are not detected and therefore go "missing". In ATLAS, there is a variable called Missing Transverse Energy or MET, that is used to represent all particles that are not detected within the ATLAS detector. This is done by adding up the square of the momenta of all the objects that are detected and taking the negative of that sum, since momentum must be conserved in the collision. In our signature, this MET variable will be used a proxy for the neutrino from the W boson. We will then apply a MET > 30 GeV selection. 

These preselections significantly reduce the number of events we need to run over to perform the analysis. This is level at which the data is now and what will be used.

#### Number of events

The number of events observed in an experiment depends on two factors:
1. How frequently the process occurs.
2. How much data have been collected.

The amount of data collected is quantified using the **integrated luminosity**. A larger integrated luminosity corresponds to more proton-proton collisions and therefore more opportunities for rare processes to occur.

Therefore, to determine the number of simulation events we should theoretically produce, we use 

$$N = \mathcal{L}\sigma$$

where $N$ is the number of events, $\mathcal{L}$ is the integrated luminosity, and $\sigma$ is the cross-section of that specific process.

##### Event weights

Imagine that two simulation samples each contain 100,000 events. Does this mean the corresponding physical processes occur equally often in nature? Not necessarily. One process may occur much more frequently than the other. The number of simulated events is chosen by physicists and does not directly reflect the true production rate of the process. Event weights allow us to convert simulation samples into realistic predictions.

In a full particle physics analysis, event weights include contributions from:

- production values (cross-sections, k-factors, filter efficiencies)
- generation weights
- Detector efficiencies
- Trigger efficiencies
- Reconstruction efficiencies
- Additional corrections

Each simulated event contributes a certain amount to the final prediction. This contribution is quantified using an event weight. The formula we'll be using is as follow:

$$\text{event weight} = \mathcal{L}\sigma\prod\epsilon_{ff}$$

These factors allow simulation to be converted into a prediction for the number of events expected in the detector.

```python
lumi = 36000.

def calc_weight(data):
    weight = (
        lumi * data["xsec"] * data["kfac"] * data["filteff"] *
        data["mcWeight"] / data["sum_of_weights"] *
        data["ScaleFactor_PILEUP"] * data["ScaleFactor_FTAG"]
    )
    el_SF = data["ScaleFactor_ELE"] * data["ScaleFactor_ElTRIGGER"] 
    muon_SF = data["ScaleFactor_MUON"] * data["ScaleFactor_MuTRIGGER"] 
    weight = ak.where(ak.firsts(data["lep_type"]) == 11, weight * el_SF, weight * muon_SF) 
    return weight
```

```python
ttbar['eventWeight'] = calc_weight(ttbar)
single_top['eventWeight'] = calc_weight(single_top)
w_jets['eventWeight'] = calc_weight(w_jets)
z_jets['eventWeight'] = calc_weight(z_jets)
diboson['eventWeight'] = calc_weight(diboson)
```

We can use our understanding of the physics to help us develop a trigger. This trigger will help us choose the "right" leptons. In this analysis, we are only looking for muons and electrons, so we keep events where a muon or an electron fired this trigger. 

```python
def cut_trig(trigE, trigM):
    # Return True if either electron or muon trigger fired
    return trigE | trigM

data = data[cut_trig(data.trigE,data.trigM)]
ttbar = ttbar[cut_trig(ttbar.trigE,ttbar.trigM)]
single_top = single_top[cut_trig(single_top.trigE,single_top.trigM)]
w_jets = w_jets[cut_trig(w_jets.trigE,w_jets.trigM)]
z_jets = z_jets[cut_trig(z_jets.trigE,z_jets.trigM)]
diboson = diboson[cut_trig(diboson.trigE,diboson.trigM)]
```

Now that we know what events fired the trigger, we now want to ensure we have the lepton in every event that fired that trigger. This is matching the lepton to the trigger. Since we could have multiple leptons in every event, we want to have at least one of those leptons matched to the trigger.

```python
def add_trigger_cut_column(dataset):
    # 1. Create an event-level boolean mask (True if event has >= 1 matched lepton)
    event_mask = ak.any(dataset["lep_isTrigMatched"], axis=1)
    
    # 2. Store it as a new column in the existing dataset
    dataset["pass_trig_cut"] = event_mask
    
    return dataset

data = add_trigger_cut_column(data)
ttbar = add_trigger_cut_column(ttbar)
single_top = add_trigger_cut_column(single_top)
w_jets = add_trigger_cut_column(w_jets)
z_jets = add_trigger_cut_column(z_jets)
diboson = add_trigger_cut_column(diboson)
```

```python
# Filter out events where no leptons were matched to trigger
data = data[data.pass_trig_cut]
ttbar = ttbar[ttbar.pass_trig_cut]
single_top = single_top[single_top.pass_trig_cut]
w_jets = w_jets[w_jets.pass_trig_cut]
z_jets = z_jets[z_jets.pass_trig_cut]
diboson = diboson[diboson.pass_trig_cut]
```
